```python
df_usa = pd.read_csv('.\data\FEDFUNDS.csv', )
df_usa.shape
```

```python
# order the date column and add date and monthly columns

years = []
months = []
for i in df_usa['DATE']:
    years.append(i[0:4])
    months.append(i[6:7])

df_usa['Year'] = years

df_usa['Month'] = months

# drop the old date column
df_usa.drop(columns='DATE', inplace=True)

df_usa 
```

```python
# make new colmun to show the Difference in Current_Rate from the previous month

df_usa['Change_From_Last_Month'] = df_usa['FEDFUNDS'].diff()

df_usa['Change_From_Last_Month'].fillna(0, inplace=True)

df_usa
```

```python
df_il = pd.read_csv('.\data\Israel interest tares.csv', encoding='ISO-8859-8')
df_il
```

```python
df_il = df_il[['TIME_PERIOD', 'OBS_VALUE']]
df_il
```

```python
# order the date column and add date and monthly columns

years = []
months = []
for i in df_il['TIME_PERIOD']:
    years.append(i[0:4])
    months.append(i[5:7] if i[5] != '0' else i[6:7])
    
df_il['Year'] = years

df_il['Month'] = months

# drop the old date column

df_il.drop(columns='TIME_PERIOD', inplace=True)

# make new colmun to show the Difference in Current_Rate from the previous month

df_il['Change_From_Last_Month'] = df_il['OBS_VALUE'].diff()

df_il['Change_From_Last_Month'].fillna(0, inplace=True)

df_il
```

```python
df_usd_il = pd.read_csv('.\\data\\USD_ILS Historical Data.csv')
df_usd_il
```

```python
df_usd_il = df_usd_il[['Date', 'Price']]
df_usd_il
```

```python
# order the date column and add date and monthly columns

years = []
months = []
for i in df_usd_il['Date']:
    years.append(i[6:10])
    months.append(i[0:2] if i[0]!= '0' else i[1:2])

df_usd_il['Year'] = years

df_usd_il['Month'] = months

# drop the old date column

df_usd_il.drop(columns='Date', inplace=True)

# make new colmun to show the Difference in Current_Rate from the previous month

df_usd_il['Change_From_Last_Month'] = df_usd_il['Price'].diff()

df_usd_il['Change_From_Last_Month'].fillna(0, inplace=True)


df_usd_il.head(30)
```
